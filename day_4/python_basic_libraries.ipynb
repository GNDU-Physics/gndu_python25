# Matplotlib 

```python
# TASK 1 - Making Sin Curve

import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(0,10,100)

y = np.sin(x) # defined a function 

plt.figure(figsize=(6,6))
plt.grid()
plt.xlabel('x -->')
plt.ylabel('Sin(x) -->')
plt.title("Sin(x) Curve")
plt.plot(x,y)

plt.show()
```

```python
# TASK 1 - Making Tan Curve

import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(0,2*np.pi,100)

y = np.tan(x) # defined a function 

plt.grid()
plt.xlabel('x -->')
plt.ylabel('Tan(x) -->')
plt.title("Tan(x) Curve")
plt.plot(x,y)

plt.style.use('classic')

plt.savefig('tanx.png', dpi=100)
plt.savefig('tanx.pdf', dpi=100)

plt.show()
```

```python
# two in one plot for sin and cos curve

x = np.linspace(0,10,100)

plt.plot(x,np.sin(x), '-g',lw=1, label="sinx")
plt.plot(x,np.cos(x), color='black',lw=5, ls=":", label="cosx") # -g,--c,-.k,:r , lw = line width and ls = line style

plt.axis("equal")

plt.legend()
plt.show()
```

```python
x = np.random.random((1,100,10))
y = np.random.random((1,100,10))
plt.scatter(x,y)
plt.show()
```

```python
# hist

data = np.random.normal(0,1,1000) # normal distribution with mean 0 and std dev = 1
plt.hist(data,bins='auto', label='histogram') # bin=int where eg bin = 20
plt.legend()
plt.show()
```

```python
# error plot

x = np.linspace(0,10,100)

y = np.sin(x) # defined a function 

error = 0.1+0.1*x
plt.errorbar(x,y,yerr=error,fmt="-o" ,color="purple", ecolor="orange" , capsize=4, elinewidth=2 )
plt.title("First Error Bar Plot")
plt.xlabel("x->>>")
plt.ylabel("sinx->>>")
plt.show()
```

### Subplots

```python
#using plt.subplot()

plt.subplot(2,1,1) # row, column, panel number 
plt.plot(x,np.sin(x),'-b')

plt.subplot(2,1,2) # row, column, panel number 
plt.plot(x,np.cos(x),':o')
plt.show()
```

```python
# object oriented using ax

x = np.linspace(0,2*np.pi,100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, axs = plt.subplots(2,1) #2x1

axs[0].plot(x,y1,'-g',label="sinx")
axs[1].plot(x,y2,'-o',label='cosx')

plt.legend()

plt.show()
```
